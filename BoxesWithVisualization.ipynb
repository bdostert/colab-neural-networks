```python
import torch
import torch.nn as nn
import collections
import numpy as np
import random
import json
import os
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from box_embeddings.parameterizations import BoxTensor
from box_embeddings.modules import BoxEmbedding
from box_embeddings.modules.volume import HardVolume, SoftVolume, BesselApproxVolume
from box_embeddings.modules.intersection import HardIntersection, GumbelIntersection
from box_embeddings.parameterizations import BoxTensor, MinDeltaBoxTensor, SigmoidBoxTensor
from box_embeddings.modules.pooling import HardIntersectionBoxPooler
from box_embeddings.modules.regularization import L2SideBoxRegularizer
from box_embeddings.parameterizations.box_tensor import BoxFactory, BoxTensor

#entities = ['grizzly_bear', 'plant', 'animal', 'snake', 'brown', 'green', 'grass', 'omnivore', 'carnivore', ]
entities = ['animal', 'mammal', 'bird', 'plant', 'dog', 'cat', 'penguin', 'parrot', 'cow', 'chicken', 'carrot', 'seed', 'food', 'sunflower',
            'rice', 'tree', 'oak', 'pine', 'polar', 'tropical', 'flower']

#trainers = [['grizzly_bear', 'animal'], ['grizzly_bear', 'brown'], ['grizzly_bear', 'omnivore'],['grass', 'plant'], ['grass', 'green'], ['grass', 'brown'], ['snake', 'animal'], ['snake', 'brown'], ['snake', 'green'], ['snake', 'carnivore']]
trainers = [['dog', 'mammal'],['cat','mammal'], ['cow','mammal'], ['bird','animal'], ['mammal', 'animal'], ['penguin','bird'], ['parrot','bird'], ['chicken','bird'],
            ['tree','plant'], ['sunflower', 'flower'], ['flower','plant'], ['oak','tree'], ['pine','tree'],['rice','plant'],['rice','food'],['chicken','food'], ['cow','food'],
            ['carrot','food'], ['oak','tropical'], ['pine','polar'], ['penguin','polar'], ['parrot','tropical']]

non_negatives = []
#Non-negatives are implications that we are not training on but which we don't want to consider as pairs to be avoided. For example, if we train on hippo implies mammal, we don't want mammal implies hippo to be a negative pair.
for a in ['penguin', 'dog', 'cat', 'cow', 'polar', 'tropical', 'parrot', 'chicken']:
    non_negatives += [[a, 'animal'], ['animal', a]]
for a in ['seed','oak', 'pine','sunflower']:
    non_negatives += [[a, 'plant'], ['plant', a]]
for pair in trainers:
    non_negatives.append([pair[1], pair[0]]) #If we train on A implies B then we don't negatively train on B implies A.
for e1 in ['polar', 'tropical']:
    for e2 in ['mammal', 'bird', 'plant', 'animal']:
        non_negatives.append([e1, e2])
        non_negatives.append([e2, e1])

negatives = []
for e1 in entities:
    for e2 in entities:
        if e1 == e2:
            continue
        if [e1, e2] not in trainers + non_negatives and [e2, e1] not in trainers + non_negatives:
            negatives.append([e1, e2])
e2ix = {}
ix2e = {}
for i, e in enumerate(entities):
    e2ix[e] = i
    ix2e[str(i)] = e
torch.set_printoptions(sci_mode=False)

epochs = 6000
print('epochs', epochs)
learning_rate = 0.5
print('learning rate', learning_rate)
embedding_dim = 2 #Try different dimensions.
print('Embedding dim', embedding_dim)
criterion = nn.MSELoss()
verbose = False
save_path = 'checkpt_dim' + '_probs_hard_intersection_train_test' + str(embedding_dim)+ 'lr' + str(learning_rate) +'.pth'
loss_log_file = 'loss_log'

class TestCE(nn.Module):
    def __init__(self):
        super(TestCE, self).__init__()
        self.embedding_dim = embedding_dim
        #The model will need to know in advance how many entities there are.
        self.hard_volume = HardVolume()
        self.soft_volume = SoftVolume()
        #self.soft_volume = SoftVolume(volume_temperature=0.5)
        #self.bessel_volume = BesselApproxVolume()
        #self.bessel_volume = BesselApproxVolume(volume_temperature=5.0, intersection_temperature=0.8)
        self.box_embedding = BoxEmbedding(len(entities), embedding_dim)
        #self.gumbel_intersection = GumbelIntersection(intersection_temperature=0.8)
        self.hard_intersection = HardIntersection()
        #self.eps = torch.tensor(0.001, requires_grad=False)
        #self.box_regularizer = L2SideBoxRegularizer(0.01, log_scale=True)
        #self.pooler = HardIntersectionBoxPooler()
        #self.logsoftmax = nn.LogSoftmax(dim=1)

    #def cross_entropy(self, pred, soft_targets): #In case we decide to use it.
    #    #return torch.mean(torch.sum(- soft_targets * self.logsoftmax(pred), 1))
    #    return torch.mean(torch.sum(- soft_targets * torch.log(pred), 1))

    def forward(self, e1, e2, predict_e1_from_e2=False):
        #The soft volume function prevents boxes that would otherwise be non-overlapping from having a zero volume of intersection.
        e1 = self.box_embedding(e1)
        e2 = self.box_embedding(e2)
        e1_vol = self.soft_volume(e1)
        e2_vol = self.soft_volume(e2)
        if predict_e1_from_e2:
            prob = self.soft_volume(self.hard_intersection(e1, e2)) - e2_vol #We subtract instead of dividing because the modules calculate log probs by default
        else:
            prob = self.soft_volume(self.hard_intersection(e1, e2)) - e1_vol
        return prob

net = TestCE()
#optimiser = torch.optim.Adam(net.parameters(), lr=learning_rate, weight_decay=1e-5)
optimiser = torch.optim.SGD(net.parameters(), lr=learning_rate, weight_decay=1e-5)
net.train()
if verbose:
    for name, param in net.named_parameters():
        if param.requires_grad:
          print(name, param.data)

if os.path.exists(save_path):
    checkpoint = torch.load(save_path)
    print('Loading checkpoint')
    net.load_state_dict(checkpoint['net_state_dict'])
    print('Checkpoint loaded')

aggregate_loss = 0
for i in range(epochs):
    e = random.choice(trainers)
    with torch.no_grad():
        e1 = torch.tensor(e2ix[e[0]]).clone().detach()
        e2 = torch.tensor(e2ix[e[1]]).clone().detach()
    #predict e2 from e1
    e_false = random.choice(negatives)
    e_false1 = torch.tensor(e2ix[e_false[0]]).clone().detach()
    e_false2 = torch.tensor(e2ix[e_false[1]]).clone().detach()
    #print('1', e1, e2, e_false)
    pred_true = net(e1, e2, False)
    pred_false = net(e_false1, e_false2, False)
    target_true = torch.tensor(1.)
    target_true = torch.unsqueeze(target_true, dim=0)
    target_true = target_true.float()
    pred_true = torch.unsqueeze(torch.exp(pred_true), dim=0)
    loss_true = criterion(pred_true, target_true)
    optimiser.zero_grad()
    loss_true.backward()
    optimiser.step()
    target_false = torch.tensor(0.)
    target_false = torch.unsqueeze(target_false, dim=0)
    target_false = target_false.float()
    pred_false = torch.unsqueeze(torch.exp(pred_false), dim=0)
    loss_false = criterion(pred_false, target_false)
    optimiser.zero_grad()
    loss_false.backward()
    optimiser.step()
    with torch.no_grad():
        aggregate_loss += loss_true + loss_false
        if i % 10 == 0:
            print('average loss', round(aggregate_loss.item() / 10, 4))
            aggregate_loss = 0
    if i > 0 and i % 100 == 0:
        torch.save({'net_state_dict': net.state_dict(),  'optimiser_state_dict': optimiser.state_dict()}, save_path)


groups  = {'trainers': trainers, 'non_negatives': non_negatives, 'negatives': negatives}
for group in list(groups):
    print('Predicted conditional probabilities of pairs in', group)
    for pair in list(sorted(groups[group])):
        with torch.no_grad():
            e1index = torch.tensor(e2ix[pair[0]])
            e2index = torch.tensor(e2ix[pair[1]])
            pred1 = torch.exp(net(e1index, e2index, False)) #Predict e2 from e1
            print('If', pair[0], 'then', pair[1], 'has prob', pred1.item())
    print()

print('Now checking implications made by an intersecting pair of entities')
print()
for pair in [['tropical', 'bird'], ['tropical', 'mammal'], ['polar', 'bird'], ['polar', 'mammal'], ['seed','food'],['bird','food'],['mammal','food']]:
    print('Checking implications of if', pair[0], '(blue)', 'and', pair[1], '(red)', 'then ...')
    with torch.no_grad():
        p1 = net.box_embedding(torch.tensor(e2ix[pair[0]]))
        p2 = net.box_embedding(torch.tensor(e2ix[pair[1]]))
        print('p1', p1.z[0], p1.z[1], p1.Z[0]-p1.z[0], p1.Z[1]-p1.z[1])
        p1p2 = net.hard_intersection(p1,p2)
        for e in entities:
            if e not in pair:
                e_emb = net.box_embedding(torch.tensor(e2ix[e]))
                vol =  net.soft_volume(net.hard_intersection(p1p2, e_emb)) - net.soft_volume(p1p2)
                print(e, '(green)', 'with prob', round(torch.exp(vol).item(), 4))
                fig, ax = plt.subplots()
                #ax.add_patch(Rectangle((0,0),1,1,fill=None, alpha=1))
                ax.set(xlim=(torch.min(torch.tensor([p1.z[0],p2.z[0],e_emb.z[0]]))-.5,torch.max(torch.tensor([p1.Z[0],p2.Z[0],e_emb.Z[0]]))+.5), ylim=(torch.min(torch.tensor([p1.z[1],p2.z[1],e_emb.z[1]]))-.5,torch.max(torch.tensor([p1.Z[1],p2.Z[1],e_emb.Z[1]])+.5)))
                #define Matplotlib figure and axis
                #add rectangle to plot
                ax.add_patch(Rectangle((p1.z[0].numpy(), p1.z[1].numpy()), (p1.Z[0]-p1.z[0]).detach().numpy(), (p1.Z[1]-p1.z[1]).detach().numpy(), edgecolor='blue', linewidth=4, linestyle='dashed', fill=None, alpha=1))
                ax.add_patch(Rectangle((p2.z[0].numpy(), p2.z[1].numpy()), (p2.Z[0]-p2.z[0]).detach().numpy(), (p2.Z[1]-p2.z[1]).detach().numpy(), edgecolor='red', linewidth=4, linestyle='dashed', fill=None, alpha=1))
                ax.add_patch(Rectangle((e_emb.z[0].numpy(), e_emb.z[1].numpy()), (e_emb.Z[0]-e_emb.z[0]).detach().numpy(), (e_emb.Z[1]-e_emb.z[1]).detach().numpy(), edgecolor='green', linewidth=4, linestyle='dashed', fill=None, alpha=1))
                plt.title('Boxes')
                plt.show()
                print()


```
